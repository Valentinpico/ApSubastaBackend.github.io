Ejercicio 29. Obtén los valores promedio de 'G1', 'G2' y 'G3' para los distintos valores posibles de 'daily_consum'

Ejercicio 30. Representa en una gráfica de barras los valores medios obtenidos en el apartado anterior. Importa los paquetes necesarios para poder representar la gráfica. Añade a la gráfica la información que consideres necesaria para entender mejor los resultados

```python

import matplotlib.pyplot as plt

# Obtener los valores promedio de G1, G2 y G3 para cada valor de daily_consum
promedios_diarios = df_math_red.groupby('daily_consum')['G1', 'G2', 'G3'].mean()

# Configurar la figura y los ejes
fig, ax = plt.subplots()

# Crear la gráfica de barras
promedios_diarios.plot(kind='bar', ax=ax)

# Personalizar la gráfica
ax.set_xlabel('daily_consum')
ax.set_ylabel('Valores promedio')
ax.set_title('Valores promedio de G1, G2 y G3 para distintos valores de daily_consum')

# Mostrar la gráfica
plt.show()
```
